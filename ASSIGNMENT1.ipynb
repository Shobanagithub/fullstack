**READING THE CSV FILE**

```r
data=read.csv("heart.csv")
print(data)
```

# ANALYZING THE DATASET

```r
print(is.data.frame(data))
print(ncol(data))
print(nrow(data))
```

# GETTING THE MAXIMUM AGE

```r
age=max(data$age)
print(age)
```

# GETTING THE MINIMUM AGE

```r
age=min(data$age)
print(age)
```

# MEAN,MEDIAN,MODE,VARIANCE,STANDARD DEVIATION

```r
mean(data$age)
```

```r
median(data$age)
```

```r
mode=function(){
  return(sort(-table(data$age))[1])
}
mode()
```

```r
var(data$age)
```

```r
sd(data$age)
```

```r
mean(data$sex)
```

```r
median(data$sex)
```

```r
mode=function(){
  return(sort(-table(data$sex))[1])
}
mode()
```

```r
var(data$sex)
```

```r
sd(data$sex)
```

# SKEWNESS

```r
install.packages("moments")
library(moments)
print(skewness(data$age))
```

```r
print(skewness(data$sex))
```

```r
colors = c("green", "orange", "brown")
months <- c("Mar", "Apr", "May", "Jun", "Jul")
regions <- c("East", "West", "North")
Values <- matrix(c(2, 9, 3, 11, 9, 4, 8, 7, 3, 12, 5, 2, 8, 10, 11),nrow = 3, ncol = 5, byrow = TRUE)
barplot(Values, main = "Total Revenue", names.arg = months,xlab = "Month", ylab = "Revenue",
col = colors, beside = TRUE)
legend("topleft", regions, cex = 0.7, fill = colors)
```

```r
colors = c("green","orange","brown")
months <- c("Mar","Apr","May","Jun","Jul")
regions <- c("East","West","North")
Values <- matrix(c(2,9,3,11,9,4,8,7,3,12,5,2,8,10,11), nrow = 3, ncol = 5, byrow = TRUE)
barplot(Values, main = "total revenue", names.arg = months, xlab = "month", ylab = "revenue", col = colors)
legend("topleft", regions, cex = 1.3, fill = colors)
```
